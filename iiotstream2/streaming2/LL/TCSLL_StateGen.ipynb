```python
import pandas as pd
import os
import sys
import time
import matplotlib.pyplot as plt
import scipy.signal
import numpy as np
import seaborn as sns
import datetime
```

```python
raw_ll=pd.read_csv('LL_IMU11.csv', usecols=['time', 'ax', 'az'])
```

```python
raw_ll['time'] =  pd.to_datetime(raw_ll['time'])
raw_ll=raw_ll.rename(columns={'time':'timestamp'})
raw_ll['ist_time']=raw_ll['timestamp']+datetime.timedelta(hours=5, minutes=30)
```

```python
raw_ll
```

```python
%matplotlib
```

```python
raw_ll["acc"] = ( raw_ll["ax"]**2 + raw_ll["az"]**2 ) ** 0.5
```

```python
# plt.plot(raw_lul['time'],raw_lul['acc'])
```

```python
acc=raw_ll[["timestamp", "acc"]]
acc.set_index('timestamp')
test_ll=acc
```

```python
test_ll
```

```python
test_ll['rlg_acc']=pd.Series.to_frame(test_ll['acc'].rolling(150, center=True).std())
```

```python
test_ll
```

```python
timethresh=15
```

```python
import scipy.signal
import numpy as np
```

```python
# plt.plot(test_ll['timestamp'],test_ll['rlg_acc'])
# plt.plot(test_ll['timestamp'],test_ll['acc']/10,'orange')
```

```python
board=scipy.signal.find_peaks(test_ll.rlg_acc, height=(0.04), distance=timethresh*100, width=1)
```

```python
board
```

```python
# plt.plot(raw_lul['time'],raw_lul['acc'])
```

```python
differ=np.diff(board[0])
differ=differ.tolist()
```

```python
differ
```

```python
for x in board[0]:
    test_ll.at[x, 'state'] = 1

ll_boards=test_ll[test_ll['state']==1]
print('Boards detected in LL:', ll_boards.shape[0])
```

```python
fig, ax1 = plt.subplots()
# ax1.plot(test_ll['timestamp'], np.ones(test_ll.shape[0])*0.01, 'green')
ax1.plot(test_ll['timestamp'], test_ll['rlg_acc'], 'orange')

# ax1.plot(test_ld['timestamp'], np.ones(test_ld.shape[0]*0.01), 'green')
# ax1.xaxis.set_major_locator(mdates.DateFormatter('%H:%M'))

ax1.set_xlabel('Time')

ax2 = ax1.twinx()
ax2.stem(ll_boards['timestamp'], ll_boards['state'], 'b:')

plt.title('Board Detections')
```
